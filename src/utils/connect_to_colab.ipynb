```python
import os
import sys
from google.colab import drive

def setup_colab_environment():
    """
    Mounts Drive, sets up project paths, creates folders, handles repo, and symlinks.
    Returns the VM project path.
    """
    try:
        drive.mount('/content/drive', force_remount=True)
        print("Google Drive mounted successfully.")
    except Exception as e:
        print("Drive mount failed:", e)
        print("Continuing without Drive persistence...")

    DRIVE_BASE = '/content/drive/MyDrive/ColabNotebooks/document-classifier'
    VM_PROJECT = '/content/document-classifier'

    if os.path.ismount('/content/drive'):
        os.makedirs(f'{DRIVE_BASE}/data', exist_ok=True)
        os.makedirs(f'{DRIVE_BASE}/models/trained_models', exist_ok=True)
        print("Drive folders ready:", DRIVE_BASE)

    if not os.path.exists(VM_PROJECT):
        print("Cloning repo...")
        !git clone https://github.com/Zeckaris/document-classifier.git {VM_PROJECT}
    else:
        print("Repo exists → pulling latest...")
        !cd {VM_PROJECT} && git pull

    os.makedirs(f'{VM_PROJECT}/src/models', exist_ok=True)

    if os.path.ismount('/content/drive'):
        data_link = f'{VM_PROJECT}/data'
        models_link = f'{VM_PROJECT}/src/models/trained_models'

        if not os.path.islink(data_link) and not os.path.exists(data_link):
            !ln -s {DRIVE_BASE}/data {data_link}
            print("Symlink created: data → Drive/data")

        if not os.path.islink(models_link) and not os.path.exists(models_link):
            !ln -s {DRIVE_BASE}/models/trained_models {models_link}
            print("Symlink created: trained_models → Drive/models/trained_models")
    else:
        print("No Drive mount → symlinks skipped (data/models will be VM-local only)")

    if VM_PROJECT not in sys.path:
        sys.path.append(VM_PROJECT)

    print("\nSetup complete!")
    print(f"Project VM path: {VM_PROJECT}")
    if os.path.ismount('/content/drive'):
        print(f"Drive base path: {DRIVE_BASE}")
    
    return VM_PROJECT


def get_project_path():
    return '/content/document-classifier'


def get_drive_path():
    return '/content/drive/MyDrive/ColabNotebooks/document-classifier'


vm_path = setup_colab_environment()
print(f"\nReady to use: vm_path = {vm_path}")
```
